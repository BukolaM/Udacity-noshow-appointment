## Data Cleaning 
In data cleaning, there are basic things to consider before proceeding with the analysis of the data. This is to  remove any inconsistencies, irregularities and to ensure that your data is of high relevance, high intergity and also for validity.
There are several data cleaning processes that would be adopted through out this dataset 

1. Renaming:Some columns headers in this dataset are misspelt  with each other, this would be looked into properly and corrected as necessary, this would give clarity and also makes it easier to understand the use case of each columns as these column holds the properties of the data

2. Drop Columns: Some columns in this dataset would be dropped due to their irrelevancy towards  further analysis. These include `patient_id` and`appointment_id`.

3. Converting our datetimes in the dataset into python datetimes:  This is to enable us do further analysis relating to a particular period within the data set.

4. Categorizing the dataset: Specific columns like age group were categorised using bin into different buckets. This  would enable us  perform specific analysis based on age range. 

5. Remove Outliers: There are certain outliers in some of the columns, such as in `age` and `handicap` columns. In this dataset, they are problematic and affect the result of our analysis.
        



#### 1. Renaming Columns

```python
#renaming the dataset columns for accuracy
df2.rename(columns = {'PatientId':'patient_id','AppointmentID':'appointment_id','ScheduledDay':'scheduled_date','Gender':'gender','AppointmentDay':'appointment_date', 'Age':'age', 'Neighbourhood':'neighbourhood','Scholarship':'scholarship',
                      'Hipertension':'hypertension', 'Diabetes':'diabetes','Alcoholism':'alcoholism','Handcap':'handicap','SMS_received':'sms_received','No-show':'no_show'}, inplace = True)
df2.head(5)
```

#### 2. Dropping Columns

```python
df2.drop(['patient_id', 'appointment_id'], axis=1, inplace=True, errors='ignore')
```

#### 3. Normalizing the scheduled_date so as to extract the years, months and date for further analysis



```python
#extraction analysis on scheduled_date
df2['new_scheduled_date'] = pd.to_datetime(df2.scheduled_date).dt.strftime('%Y-%m-%d')
df2['scheduled_month'] = pd.to_datetime(df2.scheduled_date).dt.month
df2['scheduled_year'] = pd.to_datetime(df2.scheduled_date).dt.year
df2['scheduled_day'] = pd.to_datetime(df2.scheduled_date).dt.day
df2.head()
```

#### 4. Normalizing the appointment_date so as to extract the years, months and date for further analysis

```python
#extraction analysis appointment_date
df2['new_appointment_date'] = pd.to_datetime(df2.appointment_date).dt.strftime('%Y-%m-%d')
df2['appointment_day'] = pd.to_datetime(df2.appointment_date).dt.day
df2['appointment_month'] = pd.to_datetime(df2.appointment_date).dt.month
df2['appointment_year'] = pd.to_datetime(df2.appointment_date).dt.year
df2['appointment_day_of_week'] = pd.to_datetime(df2.appointment_date).dt.day_name()
df2.head()
```

#### 5. Normalizing the appointment_date so as to extract the actual month value

```python
# Transforming appointment_date to actual month value
from datetime import datetime

def get_month(month):
    
    return datetime.strptime(month, "%Y-%m-%dT%H:%M:%S%z").strftime("%B")

    
df2["appointment_month_value"] = df2["appointment_date"].apply(get_month)

df2["appointment_month_value"].unique()
```

#### 6. Categorizing the dataset into different age ranges

```python
#grouping age into different age groups 
import numpy as np
bins = [0, 5, 13, 18, 35, 65, np.inf]
names = ['<5','5-13','13-18','18-35','35-65', '65+']
df2['age_range'] = pd.cut(df2['age'], bins, labels=names)
df2.head()
```

#### 7. Removing the outliers in age column

```python
# removing outliers i.e from 0 downwards as age cannot be less than 0
df2 = df2[df2['age'] >= 0]
df2.head()
```

#### 8. Removing the outliers in handicap column

```python
# removing handicap outliers, handicap can only either be a yes or no which is also a 1 or 0, therefore, we remove outliers from above 1
df2 = df2[df2['handicap'] < 2]
df2.head()
```

## Exporting the Cleaned Data to a New csv 

```python
#export the data to a new csv file
df3 = df2.to_csv('./noshowappointments-clean_data.csv', index = False)
```

 # Data Exploratory Analysis 
 

```python
#using pandas to read the new csv file
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#import the saved clean data
df3 = pd.read_csv('./noshowappointments-clean_data.csv',sep=',')
df3.head()
```

```python
df3.info()
```

### <span style='color:blue'> Question 1. How frequently do patients show up for their appointments during the week?</span>

```python
# Creating an index for appointment day of the week to enable us determine if weekdays influences a patient to show up
appointment_index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

#ordering the index 
df3['appointment_day_of_week'] = pd.Categorical(df3['appointment_day_of_week'], 
                                categories=appointment_index,ordered=True)


#Grouping based on appointment_week and  and appointment, this is to enable us compare both columns 
weekly_appointment = df3[['appointment_day_of_week','no_show']].groupby(['appointment_day_of_week',
                                                            'no_show']).size().reset_index(name='counts')

weekly_appointment_pivot = (
    weekly_appointment.pivot(
    values="counts",
    index="appointment_day_of_week",
    columns="no_show",
    ).sort_values('appointment_day_of_week') 
    )

#using bar chart to plot the visuals
weekly_appointment_pivot.plot(kind="bar")
plt.title('corresponding days of the week to appointment bookings') 
plt.xlabel("days of the week")
plt.ylabel("total number of patients")
```

#### From the bar chart above, this shows and reflects that its more convenient for patients to show up for their appointment at best in the early days of the week while weekends has the least show up.

### <span style='color:blue'> Question 2. How does the neighborhood affect the rate at which people show up for their appointments?</span>


```python
#Grouping based on neighbourhood  and appointment
neighbourhood= df3[['neighbourhood','no_show']].groupby(['neighbourhood','no_show']).size().reset_index(name='counts')
neighbourhood

neighbourhood_pivot = (
  neighbourhood.pivot(
  values="counts",
  index="neighbourhood",
  columns="no_show",
  ).sort_values('neighbourhood')
)
neighbourhood_pivot.head()
```

```python
#plotting details using a stacked chart

plot = neighbourhood_pivot.plot.barh(stacked=True, figsize = (50,90))
for container in plot.containers:
    plot.bar_label(container,fontsize = 25)
plt.title('neighbourhood distibution by no_show',fontsize = 40) 
plt.xlabel("No_Show",fontsize = 25)
plt.ylabel("neighbourhood",fontsize = 40)
```

#### Based on the data , it can be seen that there is no significant relationship between the neighbourhood and how people turn up for their appointment however, JARDIM DA PENHA neighbourhood has the highest number of patients that showed up for their appointment.


### <span style='color:blue'> Question 3. Does SMS reminders influence how patients respond to their appointments? </span>


```python
#Grouping based on SMS_received and  and appointment
sms_counts= df3[['sms_received','no_show']].groupby(['sms_received','no_show']).size().reset_index(name='counts')
sms_counts

#comparing the columns using a pivot
sms_count_pivot = (
    sms_counts.pivot(
    values="counts",
    index="sms_received",
    columns="no_show",
    ).sort_values('sms_received')
)
sms_count_pivot
```

```python
#plotting the result using a barh chart
sms_count_pivot.plot(kind="barh", figsize = (10,3))
plt.title('no show by SMS received') 
plt.xlabel("total number of patient")
plt.ylabel("sms received")
```

#### The chart above indicates that sms reminders does not influence patients showing up for their appointment,  as the total number of patients without sms reminders exceeded those with sms reminders.

### <span style='color:blue'>Question 4. Are the handicapped patients able to meet or miss their appointments? </span>


```python
#Grouping based on handicap and  and appointment
handicap_count = df3[['handicap','no_show']].groupby(['handicap','no_show']).size().reset_index(name='counts')
handicap_count

handicap_count_pivot = (
    handicap_count.pivot(
    values="counts",
    index="handicap",
    columns="no_show",
    ).sort_values('handicap')
)
handicap_count_pivot
```

```python
#plotting the result using a bar chart
handicap_count_pivot.plot(kind="bar")
plt.title('comparison between handicap and appointment') 
plt.xlabel('handicap status')
plt.ylabel('total number of patients')
```

#### It can be deduced from the chart above that only few handicap patients showed up for their appointment.

###  <span style='color:blue'>Question 5. How do hypertension, diabetes, and alcoholism affect the ability of patients to show up or not for their appointments?</span>


```python
#Grouping based on hypertension,diabetes,alcoholism  and appointment
other_health_factors= df3[['hypertension','diabetes','alcoholism',
                    'no_show']].groupby(['hypertension','diabetes','alcoholism',
                    'no_show']).size().reset_index(name='counts')

health_factors= (
    other_health_factors.pivot(
    values="counts",
    index=['hypertension','diabetes','alcoholism'],
    columns="no_show"
    )
    )
health_factors
```

```python
#plotting the result using a barh chart
health_factors.plot(kind="bar", figsize = (10,4))
plt.title('comparison between hypertension, diabetes and alcoholism by appointment') 
plt.xlabel('hypertension, diabetes and alcoholism in that order')
plt.ylabel('total number of patients')
```

#### The above chart indicates that patients without all the status show up the highest for their appointment, while, patients with these three health conditions show up the least. However, it can also deduced that hypertensive patients always show up for their appointments the most.

###  <span style='color:blue'>Question 6. What is the percentage of scholarship recipients? Does this affect whether they show up for their appointment or not?</span>

```python
#Grouping based scholarship and appointment
df3[['scholarship','no_show']].groupby(['no_show']).sum().plot(kind='pie', y='scholarship', autopct='%1.0f%%', figsize=(6,6),
colors = ['orange', 'steelblue'], title='Percentage of scholarship by appointment')
```

#### The pie chart  shows that patients with the scholarship benefits show up highest for their appointment.

###  <span style='color:blue'>Question 7. Does age group have any impact on a patient's appointment availability?</span>

```python
# Creating an index for age_range to enable in an ascending order
age_range_index = ['<5', '5-13','13-18', '18-35', '35-65', '65+']
# Ordering the index
df3['age_range'] = pd.Categorical(df3['age_range'], categories=age_range_index, ordered=True)

#Grouping based on age_range and appointment, this is to enable us compare both columns and see if a particular age group turn up than the other
agegroup_count= df3[['age_range','no_show']].groupby(['age_range','no_show']).size().reset_index(name='counts')
agegroup_count

#comparing the columns using a pivot
agegroup_count_pivot = (
    agegroup_count.pivot(
    values="counts",
    index="age_range",
    columns="no_show",
    ).sort_values('age_range')
    )
agegroup_count_pivot
```

```python
#plotting the result using a bar chart
agegroup_count_pivot.plot(kind="bar", figsize=(10,4))
plt.title('age range by appointment') 
plt.xlabel("age range")
plt.ylabel(" total number of patients")
```

#### From the bar chart above, it can be seen the range range between 35-65 has the highest appointment show. 

###  <span style='color:blue'>Question 8. Is there any relationship between hypertension and age among the patients?</span>

```python
#grouping age_range by hypertension
df3['has_hypertension'] = np.where(df3['hypertension']== 1, 'yes', 'no')
age_h_count= df3[['age_range','has_hypertension']].groupby(['age_range',
                'has_hypertension']).size().reset_index(name='counts')

#comparing the columns using a pivot
age_h_pivot = (
    age_h_count.pivot(
    values="counts",
    index=["has_hypertension"],
    columns="age_range",
    )
    )
#plotting the result using a bar chart
plot = age_h_pivot.plot(kind='bar', figsize=(12,8))
for container in plot.containers:
    plot.bar_label(container)
plt.title('age range by hypertention status') 
plt.xlabel("age ranges")
plt.ylabel("total number of patients")
```

#### From the bar chart above, it can be seen the range range between 35-65 has the highest number of patients without hypertension , this can be as well deduced from them having the highest appointment show up as analysed earlier.

###  <span style='color:blue'>Question 9. Is there any relationship between alcoholism and age among the patients?</span>

```python
#grouping age_range by alcoholism to see if its a contributor for the some patients which causes low turn up
df3['h_alcoholism'] = np.where(df3['alcoholism']== 1, 'yes', 'no')
age_h_alcoholism= df3[['age_range','h_alcoholism']].groupby(['age_range',
                'h_alcoholism']).size().reset_index(name='counts')

#comparing the columns using a pivot
age_alcoholism_pivot = (
    age_h_alcoholism.pivot(
    values="counts",
    index=["h_alcoholism"],
    columns="age_range",
    )
    )
#plotting the result using a bar chart
plot = age_alcoholism_pivot.plot(kind='bar', figsize=(12,8))
for container in plot.containers:
    plot.bar_label(container)
plt.title('age range by alcoholism status') 
plt.xlabel("age ranges")
plt.ylabel("total number of patients")
```

#### From the bar chart above, it can be seen the range range between 35-65 has the highest number of patients without alcoholism, same reason as earlier stated.

###  <span style='color:blue'>Question 10. Is there any relationship between diabetes and age among the patients?</span>

```python
#grouping age_range by diabetes to see if its a contributor for the some patients which causes low turn up
df3['h_diabetes'] = np.where(df3['diabetes']== 1, 'yes', 'no')
age_h_diabetes= df3[['age_range','h_diabetes']].groupby(['age_range',
                'h_diabetes']).size().reset_index(name='counts')

#comparing the columns using a pivot
age_diabetes_pivot = (
  age_h_diabetes.pivot(
    values="counts",
    index=["h_diabetes"],
    columns="age_range",
    )
    )
#plotting the result using a bar chart
plot = age_diabetes_pivot.plot(kind='bar', figsize=(12,4))
for container in plot.containers:
    plot.bar_label(container)
plt.title('age range by diabetes status') 
plt.xlabel("age ranges")
plt.ylabel("total number of patients")
```

#### From the bar chart above, it can be seen the range range between 35-65 has the highest number of patients without diabtes, same reason as earlier stated.

###  <span style='color:blue'> Question 11. Are patients given scholarship based on gender?</span>

```python
#grouping gender by scholarship
gender_scholarship_count= df3[['gender','scholarship']].groupby(['gender','scholarship',]).size().reset_index(name='counts')
gender_scholarship_count
```

```python
#to determine scholarhip based on gender
df3[['scholarship','gender']].groupby(['gender']).sum().plot(kind='pie', y='scholarship',figsize=(6,6), autopct='%1.0f%%',
colors = ['red', 'steelblue'],title='Scholarship_gender')
```

#### The pie chart above indicates that scholarship is not on an equal basis, the female patients enjoy the scholarship opportunities higher than the male patients.

###  <span style='color:blue'>Question 12. What is the monthly analysis of show-up and no-show appointments?</span>

```python
#Grouping based on appointment_month and  and appointment, this is to enable us compare both columns 
monthly_appointment = df3[['appointment_month_value','no_show']].groupby(['appointment_month_value',
                                                            'no_show']).size().reset_index(name='counts')
monthly_appointment_pivot = (
   monthly_appointment.pivot(
    values="counts",
    index="appointment_month_value",
    columns="no_show",
    ).sort_values('appointment_month_value') 
    )

#using bar chart to plot the visuals
monthly_appointment_pivot.plot(kind="bar")
plt.title('corresponding month to appointment bookings') 
plt.xlabel("months")
plt.ylabel("total number of patients")
```

#### It can be deduced from the chart above that patients show up for their appointment mostly in May while April had the least show up

# Conclusions

From the overall analysis, certain conclusions have been drawn, which include the following:

Most of the patients were generally relatively satisfied with the weekday booking of appointments, while weekends' appointments were significantly lower.

Patients on the scholarship showed up for appointments at a higher rate than those without scholarship benefits. It was also observed that the scholarship was not evenly distributed among both genders, as females were seen to be on scholarship with 82%, while the male gender was at 18%. My recommendation here would be to ensure that the scholarship is evenly distributed across both genders.

It was also observed that SMS reminders have no effect on whether patients attend their appointments, as the data showed that a lot of patients who got sms still did not attend the appointment.

We also went further to analyze other parameters, which are hypertension, alcoholism, diabetes, and a patient's being handicapped. It was generally observed that patients who showed positive trends towards these parameters were less likely to attend an appointment. However, patients who indicated they had none of these conditions were more likely to attend their appointments. This negative correlation against this condition points to the fact that patients who have not detected if they have any health conditions have a high tendency to want to determine their current state of health and will attend their appointment.

My recommendation here will be to increase awareness among patients who have indicated that they have a health condition on why it is still important for them to attend their appointment.

Finally, we also observed that patients in the age range of 35 and 65 are more likely to attend their appointments, followed by those aged 18 to 35, and then 65 and above.
